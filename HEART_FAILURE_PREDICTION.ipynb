```python
# Import libraries
import numpy as np
import pandas as pd
```

```python
# Load the dataset
hfp = pd.read_csv("/content/heart_failure_data.csv")
hfp
```

```python
# Split the dataset
x = hfp.drop(['Target Label'], axis = 1)
y = hfp['Target Label']
```

```python
# Checking the independent cols
x
```

```python
# Checking the dependent cols
y
```

```python
# Split the dataset into train and test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x , y, test_size = 0.2, random_state = 42)
```

```python
# Chceking the x_train
x_train
```

```python
# Checking the x_test
x_test
```

```python
# Checking the y_train
y_train
```

```python
# Checking the y_test
y_test
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(x_train)
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)
```

```python
# Training
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()

# Fitting the model into training data
rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_test)
print(y_pred)
```

```python
# Evaluation
score = rfc.score(x_test, y_test)
print(score)
```

```python
# To get classification report
from sklearn.metrics import classification_report

#
print(classification_report(y_test, y_pred))
```

```python
# To get confusion metrix via heatmap

import seaborn as sns
from sklearn.metrics import confusion_matrix
classes_names = ['class 1', 'class 2']
cm = pd.DataFrame(confusion_matrix(y_test, y_pred), columns = classes_names, index = classes_names)

# Visualisation
sns.heatmap(cm, annot = True, fmt = 'd')
```
